Conflict between humans and tigers threatens tiger populations, kills people, and reduces public support for conservation. 

Researchers investigated causes of human deaths by tigers near the protected area of Chitwan National Park, Nepal. Eighty-eight people were killed by 36 individual tigers between 1979 and 2006, mainly within 1 km of the park edge. Such information may be helpful to identify activities that increase vulnerability to attack.

The file [Datasets/tiger deaths.csv](/edit/Variation1/Self-study%20Notebooks/Datasets/tiger%20deaths.csv) contains the main activities of the 88 people at the time they were attacked and killed. 

1. Import the pandas and seaborn modules.
2. Read in the csv file [Datasets/tiger deaths.csv](/edit/Variation1/Self-study%20Notebooks/Datasets/tiger%20deaths.csv) into a pandas DataFrame called `tigers`. 
3. Plot a bar plot of `tigers['activity']`and add some useful annotation.

````python
import pandas as pd
import seaborn as sns

tiger = pd.read_csv('Datasets/tiger deaths.csv')

g = sns.displot(y=tiger['activity'], shrink=0.8)

g.ax.set_ylabel('Activity at time of death')
g.ax.set_xlabel('Number of deaths')
g.ax.set_title('Activity of 88 people at time of tiger attack');
````
